```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.linspace(-np.pi,10,1000)
```

```python
C,S = np.cos(X), np.sin(X)
```

```python
plt.figure(figsize=(9,3),dpi=80)
plt.plot(X,C,c='g',linewidth=2.0, linestyle="-",label="Cosine")
plt.plot(X,S,c='b',linewidth=2.0, linestyle="-",label="Sine")
plt.xlabel("X-values")
plt.ylabel("Value of C and S")
plt.legend()
plt.show()
```
